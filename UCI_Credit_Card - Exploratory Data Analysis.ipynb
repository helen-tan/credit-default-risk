```python
import pandas as pd

df = pd.read_csv("UCI_Credit_Card.csv")
print(df)
```

# Data Cleaning

```python
df.info()
```

```python
# Remove null values if any
df = df.dropna()
print(df)
```

```python
# Rename column name 'PAY_0' to 'PAY_1'

df.rename(columns={'PAY_0':'PAY_1'}, inplace=True)
print(df)
```

```python
df.describe()
```

```python
# In Education column, change 0 and 6 to 5

df.loc[:,'EDUCATION'] = df.loc[:,'EDUCATION'].replace(0,5)
df.loc[:,'EDUCATION'] = df.loc[:,'EDUCATION'].replace(6,5)


condition1 = df['EDUCATION'] == 0
condition2 = df['EDUCATION'] == 6

print(df[condition1 | condition2])
# If no records are printed, then it means 0 and 6 have been successfully replaced to 5
```

```python
df.describe()
```

```python
# change 0, -1, -2 to 0 for all PAY_X values

#change = ['PAY_1', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']

#for i in change:
    #df[df[i]<0] = 0

df.loc[:,'PAY_1'] = df.loc[:,'PAY_1'].replace(-1,0)
df.loc[:,'PAY_1'] = df.loc[:,'PAY_1'].replace(-2,0)

df.loc[:,'PAY_2'] = df.loc[:,'PAY_2'].replace(-1,0)
df.loc[:,'PAY_2'] = df.loc[:,'PAY_2'].replace(-2,0)

df.loc[:,'PAY_3'] = df.loc[:,'PAY_3'].replace(-1,0)
df.loc[:,'PAY_3'] = df.loc[:,'PAY_3'].replace(-2,0)

df.loc[:,'PAY_4'] = df.loc[:,'PAY_4'].replace(-1,0)
df.loc[:,'PAY_4'] = df.loc[:,'PAY_4'].replace(-2,0)

df.loc[:,'PAY_5'] = df.loc[:,'PAY_5'].replace(-1,0)
df.loc[:,'PAY_5'] = df.loc[:,'PAY_5'].replace(-2,0)

df.loc[:,'PAY_6'] = df.loc[:,'PAY_6'].replace(-1,0)
df.loc[:,'PAY_6'] = df.loc[:,'PAY_6'].replace(-2,0)


condition3 = df['PAY_1'] == -1
condition4 = df['PAY_1'] == -2

condition5 = df['PAY_2'] == -1
condition6 = df['PAY_2'] == -2

condition7 = df['PAY_3'] == -1
condition8 = df['PAY_3'] == -2

condition9 = df['PAY_4'] == -1
condition10 = df['PAY_4'] == -2

condition11 = df['PAY_5'] == -1
condition12 = df['PAY_5'] == -2

condition13 = df['PAY_6'] == -1
condition14 = df['PAY_6'] == -2

print(df[condition3 | condition4 | condition5| condition6 | condition7 | condition8 | condition9 | condition10 | condition11 | condition12 | condition13 | condition14])
# If no records are printed, then it means -1 and -2 have been successfully replaced to 0
```

```python
print(df)
```

```python
import seaborn as sns

ax = sns.countplot(x="default.payment.next.month", data=df).set(title="Count of Non-Defaults and Defaults")
```

```python
condition_no_default = df["default.payment.next.month"] == 0
condition_default = df["default.payment.next.month"] == 1


print('Total no. of people who did not default:')
print(df[condition_no_default].count())

print('Total no. of people who defaulted:')
print(df[condition_default].count() )

# Total number of non-defaults = 23364
# Total number of defaults = 6636
```

# Age vs Default Status

```python
# Age vs Default count 

import seaborn as sns

sns.boxplot(df.default_payment_next_month, df.AGE).set(title="Age vs Count of Non-defaults and Defaults")
```

```python
# Info on Age of people who did not default

condition_no_default = df["default_payment_next_month"] == 0

df_no_default = df[condition_no_default]

print("Info on Age of people who did not default")
df_no_default["AGE"].describe()
```

```python
# Info on Age of people who defaulted

condition_default = df["default_payment_next_month"] == 1

df_default = df[condition_default]
print("Info on Age of people who defaulted")
df_default["AGE"].describe()
```

```python
df["AGE"].describe()
```

```python
# Create 8 bins for AGE, with a bin interval of (79 - 21)/8 = 7.25 = ~8 (round up)

df['AgeBin'] = pd.cut(df['AGE'],[21, 29, 37, 45, 53, 61, 69, 77, 85])
print(df['AgeBin'].value_counts())
```

```python
df['default_payment_next_month'].groupby(df['AgeBin']).value_counts(normalize = True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))

df['AgeBin'] = df['AgeBin'].astype('str')
AgeBin_order = ['(21, 29]', '(29, 37]', '(37, 45]', '(45, 53]',
                '(53, 61]', '(61, 69]', '(69, 77]', '(77, 85]']


ax = sns.countplot(data = df, x = 'AgeBin', hue="default_payment_next_month", palette = "rocket", order = AgeBin_order)


plt.xlabel("Age Group", fontsize= 12)
plt.ylabel("No. of Clients", fontsize= 12)
plt.ylim(0,10000)
ax.tick_params(axis="x", labelsize= 10)

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.075, p.get_height()+300))


plt.show()

```

```python
# Info on credit amount taken by 
# people who defaulted

condition_no_default = df["default_payment_next_month"] == 1

df_no_default = df[condition_no_default]

print("Info on Credit Amount of people who defaulted")
df_no_default["LIMIT_BAL"].describe()
```

```python
df["LIMIT_BAL"].describe()
```

```python
# Create 10 bins for LIMIT_BAL, with a bin interval of (1,000,000 - 1,000)/10 = 99,900

#df['LimitBin'] = pd.cut(df['LIMIT_BAL'],[1000, 100900, 200800, 300700, 400600, 500500, 600400, 700300, 800200, 900100, 1000000])
#print(df['LimitBin'].value_counts())


# Create 8 bins for LIMIT_BAL, with a bin interval of (1,000,000 - 1,000)/8 = 124,875

df['LimitBin'] = pd.cut(df['LIMIT_BAL'],[1000, 125875, 250750, 375625, 500500, 625375, 750250, 875125, 1000000])
print(df['LimitBin'].value_counts())
```

```python
df['default_payment_next_month'].groupby(df['LimitBin']).value_counts(normalize = True)
```

```python
df['LimitBin'] = df['LimitBin'].astype('str')
LimitBin_order = ['(1000, 125875]', '(125875, 250750]', '(250750, 375625]', '(375625, 500500]',
                '(500500, 625375]', '(625375, 750250]', '(750250, 875125]', '(875125, 1000000]']

plt.figure(figsize=(12,4))

ax = sns.countplot(data = df, x = 'LimitBin', hue="default_payment_next_month", palette = "rocket", order = LimitBin_order)


plt.xlabel("Amount of Given Credit", fontsize= 12)
plt.ylabel("No. of Clients", fontsize= 12)
plt.ylim(0,11000)
ax.tick_params(axis="x", labelsize= 8.5)

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.075, p.get_height()+300))


plt.show()
```

```python
plt.figure(figsize=(14,4))

ax = sns.barplot(x = "LimitBin", y = "default_payment_next_month", data = df, palette = 'rocket', ci = None, order = LimitBin_order)

plt.xlabel("Amount of Given Credit", fontsize= 12)
plt.ylabel("% of Default", fontsize= 12)
plt.ylim(0,0.5)

for p in ax.patches:
    ax.annotate("%.2f" %(p.get_height()), (p.get_x()+0.25, p.get_height()+0.03),fontsize=13)

plt.show()
```
